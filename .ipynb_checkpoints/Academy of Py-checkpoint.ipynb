```python
# Retrieve pandas library and access operating system
import pandas as pd
import csv
import os
import numpy as np

# Name file path w/ variable
schools_csv = os.path.join('Resources/schools data.csv')
students_csv = os.path.join('Resources/students data.csv')

# Read csv files
schools_df = pd.read_csv(schools_csv)
students_df = pd.read_csv(students_csv)

schools_df.head()
students_df.head()




```

```python
#District Summary

# Total number of schools
total_schools = len(schools_df["school_name"])


# Toal number of students
total_students = len(students_df["student_name"])


# Total budget for all 15 schools
total_budget = schools_df["budget"].sum()

# Average math score for all schools
average_math = students_df["math_score"].mean()


# Average reading score for all schools
average_reading = students_df["reading_score"].mean()


# Overall average combined score
overall_passing_rate = (average_math + average_reading)/2



# Percent of students with a 70% or greater math score
students_passing_math = students_df.loc[students_df["math_score"] >= 70]["math_score"].count()
percent_passing_math = students_passing_math/total_students


# Percent of students with a 70% or greater reading score
students_passing_reading = students_df.loc[students_df["reading_score"] >= 70]["reading_score"].count()
percent_passing_reading = students_passing_reading/total_students


# Data Frame with results
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Percent Passing Math":[percent_passing_math],
    "Average Reading Score": [average_reading],
    "Percent Passing Reading": [percent_passing_reading],
    "Overall Passing Rate": [overall_passing_rate]
})


# Format Data Frame
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].astype(float).map("{:.2f}".format)
district_summary["Percent Passing Math"] = district_summary["Percent Passing Math"].astype(float).map("{:.1%}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].astype(float).map("{:.2f}".format)
district_summary["Percent Passing Reading"] = district_summary["Percent Passing Reading"].astype(float).map("{:.1%}".format)
district_summary["Overall Passing Rate"] = district_summary["Overall Passing Rate"].astype(float).map("{:.1f}%".format)



district_summary

```

```python
# School Summary


# Merge the two csv files into one file
education_df = pd.merge(schools_df, students_df, on="school_name")
education_df.head()

# Extract only the needed columns
education_df = education_df.loc[:, ["school_name", "type", "size", "budget", "Student ID", "grade", "reading_score", "math_score" ]]


# Reset index
education = education_df.set_index("school_name")


# Group by school
schools_group_df = education_df.groupby(["school_name"])


# School Type
school_types = schools_df.set_index("school_name")['type']


# Total Students for each school
total_students = schools_group_df["Student ID"].count()


# Total School Budget
school_budget = schools_df.set_index("school_name")["budget"]


# Per Student Budget
budget_per_student = school_budget/total_students


# Average math score for each school
school_average_math = schools_group_df["math_score"].mean()


# Average reading score for each school
school_average_reading = schools_group_df["reading_score"].mean()


# Percent of students with a 70% or greater math score
students_passing_math_by_school = students_df[students_df["math_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_math_by_school = students_passing_math_by_school/total_students

# Percent of students with a 70% or greater reading score
students_passing_reading_by_school = students_df[students_df["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_reading_by_school = students_passing_reading_by_school/total_students


# Overall Passing Rate
overall_passing_rate_by_school = (percent_passing_math_by_school + percent_passing_reading_by_school)/2

# Data Frame with results
school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": total_students,
    "Total School Budget": school_budget,
    "Budget per Student in Dollars": budget_per_student,
    "Average Math Score": school_average_math,
    "Percent Passing Math": percent_passing_math_by_school,
    "Average Reading Score": school_average_reading,
    "Percent Passing Reading": percent_passing_reading_by_school,
    "Overall Passing Rate": overall_passing_rate_by_school
})

# Format Data Frame
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,}".format)
school_summary["Budget per Student in Dollars"] = school_summary["Budget per Student in Dollars"].astype(int)
school_summary["Average Math Score"] = school_summary["Average Math Score"].astype(float).map("{:.2f}".format)
school_summary["Percent Passing Math"] = school_summary["Percent Passing Math"].astype(float).map("{:.1%}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].astype(float).map("{:.2f}".format)
school_summary["Percent Passing Reading"] = school_summary["Percent Passing Reading"].astype(float).map("{:.1%}".format)
school_summary["Overall Passing Rate"] = school_summary["Overall Passing Rate"].astype(float).map("{:.1%}".format)


school_summary
```

```python
# Top Performing Schools
top5_performing_schools = school_summary.sort_values(["Overall Passing Rate"], ascending = False)
top5_performing_schools.head(5)

```

Top Performing Schools Analysis

This particular data frame would be of great use to a proponent of the charter school movement who is looking to persuade funders.  Yet, I find it problematic that readers will place too much power on the column reporting Overall Passing Rate.  It is a misleading figure because the value has been extrapolated.  It isn't simply reporting on raw data.  Instead, it is counting the number of students who have met an arbitrary academic goalpost out of the number of students who comprise the total school poulation.  This practice favors charter schools and therefore dramaticizes those schools as dominating in academics.

```python
# Bottom Performing Schools
bottom5_performing_schools = school_summary.sort_values(["Overall Passing Rate"], ascending = True)
bottom5_performing_schools.head(5)
```

Bottom Performing Schools Analysis

The opposite effect is present in the data frame capturing the lowest-performing schools.  If we zoom in on the difference in average math and reading scores from the "top" and "bottom" school (as determined using Overall Passing Rate as the value indicator), we see there is a difference of 6.22 average math score points and 3.24 average reading score points.  This data frame begs the question, is an overall score difference of less than 5 points on the actual test that measures performance significant enough to deem one school a top-performer and the other a bottom performer?  

```python
# Math Scores by Grade Level
nine_math = students_df.loc[students_df["grade"] == "9th"].groupby("school_name")["math_score"].mean()
ten_math = students_df.loc[students_df["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleven_math = students_df.loc[students_df["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelve_math = students_df.loc[students_df["grade"] == "12th"].groupby("school_name")["math_score"].mean()

# Data Frame with results
math_scores_by_grade = pd.DataFrame({
    "9th":  nine_math,
    "10th": ten_math,
    "11th": eleven_math,
    "12th": twelve_math 
})

# Format Date Frame
math_scores_by_grade = math_scores_by_grade[['9th', '10th', '11th', '12th']]

math_scores_by_grade["9th"] = math_scores_by_grade["9th"].map("{:.2f}".format)
math_scores_by_grade["10th"] = math_scores_by_grade["10th"].map("{:.2f}".format)
math_scores_by_grade["11th"] = math_scores_by_grade["11th"].map("{:.2f}".format)
math_scores_by_grade["12th"] = math_scores_by_grade["12th"].map("{:.2f}".format)

math_scores_by_grade

```

```python
# Reading Scores by Grade Level
nine_reading = students_df.loc[students_df["grade"] == "9th"].groupby("school_name")["reading_score"].mean()
ten_reading = students_df.loc[students_df["grade"] == "10th"].groupby("school_name")["reading_score"].mean()
eleven_reading = students_df.loc[students_df["grade"] == "11th"].groupby("school_name")["reading_score"].mean()
twelve_reading = students_df.loc[students_df["grade"] == "12th"].groupby("school_name")["reading_score"].mean()

# Data Frame with results
reading_scores_by_grade = pd.DataFrame({
    "9th":  nine_reading,
    "10th": ten_reading,
    "11th": eleven_reading,
    "12th": twelve_reading 
})

# Format Date Frame
reading_scores_by_grade = reading_scores_by_grade[['9th', '10th', '11th', '12th']]

reading_scores_by_grade["9th"] = reading_scores_by_grade["9th"].map("{:.2f}".format)
reading_scores_by_grade["10th"] = reading_scores_by_grade["10th"].map("{:.2f}".format)
reading_scores_by_grade["11th"] = reading_scores_by_grade["11th"].map("{:.2f}".format)
reading_scores_by_grade["12th"] = reading_scores_by_grade["12th"].map("{:.2f}".format)

reading_scores_by_grade
```

```python
# Scores by School Spending
school_spending_bins = [0, 585, 615, 645, 675]
group_names1 = ["< $585", "$585-615", "$615-645", "$645-675"]

# Add variable of school spending (with bins) to original data set
education_df["school_spending_bins"] = pd.cut(education_df["budget"]/education_df["size"], bins = school_spending_bins, labels = group_names1)

# Group by school spending
school_spending = education_df.groupby("school_spending_bins")

# Refigure average scores by school spending
math_by_spending = school_spending["math_score"].mean()
reading_by_spending = school_spending["reading_score"].mean()
pass_math_by_spending = education_df[students_df["math_score"] >= 70].groupby("school_spending_bins")["Student ID"].count()/school_spending["Student ID"].count()
pass_reading_by_spending = education_df[students_df["reading_score"] >= 70].groupby("school_spending_bins")["Student ID"].count()/school_spending["Student ID"].count()
pass_overall_by_spending = education_df[(students_df["math_score"] >= 70) & (education_df["reading_score"] >= 70)].groupby("school_spending_bins")["Student ID"].count()/school_spending["Student ID"].count()

# Set-up Data Frame
scores_by_school_spending = pd.DataFrame({
    "Average Math Score": math_by_spending,
    "Percent Passing Math": pass_math_by_spending,
    "Average Reading Score": reading_by_spending,
    "Percent Passing Reading": pass_reading_by_spending,
    "Overall Passing Rate": pass_overall_by_spending
})


# Insert Title for bins and index by them
scores_by_school_spending.index.name = "Budget per Student"
scores_by_school_spending = scores_by_school_spending.reindex(group_names1)

# Format Data Frame
scores_by_school_spending["Average Math Score"] = scores_by_school_spending["Average Math Score"].map("{:.2f}".format)
scores_by_school_spending["Percent Passing Math"] = scores_by_school_spending["Percent Passing Math"].map("{:.2%}".format)
scores_by_school_spending["Average Reading Score"] = scores_by_school_spending["Average Reading Score"].map("{:.2f}".format)
scores_by_school_spending["Percent Passing Reading"] = scores_by_school_spending["Percent Passing Reading"].map("{:.2%}".format)
scores_by_school_spending["Overall Passing Rate"] = scores_by_school_spending["Overall Passing Rate"].map("{:.2%}".format)


scores_by_school_spending
```

```python
# Scores by School Size
school_size_bins = [0, 1000, 2000, 5000]
group_names2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Add variable of school size (with bins) to original data set
education_df["school_size_bins"] = pd.cut(education_df["size"], bins = school_size_bins, labels = group_names2)

# Create new object with schools grouped by school size created above
by_size = education_df.groupby("school_size_bins")

# Refigure averge scores within each school-size bin
math_by_size = by_size["math_score"].mean()
reading_by_size = by_size["reading_score"].mean()
pass_math_by_size = education_df[students_df["math_score"] >= 70].groupby("school_size_bins")["Student ID"].count()/by_size["Student ID"].count()
pass_reading_by_size = education_df[students_df["reading_score"] >= 70].groupby("school_size_bins")["Student ID"].count()/by_size["Student ID"].count()
pass_overall_by_size = education_df[(students_df["math_score"] >= 70) & (education_df["reading_score"] >= 70)].groupby("school_size_bins")["Student ID"].count()/by_size["Student ID"].count()

# Set-up Data Frame
scores_by_school_size = pd.DataFrame({
    "Average Math Score": math_by_size,
    "Percent Passing Math": pass_math_by_size,
    "Average Reading Score": reading_by_size,
    "Percent Passing Reading": pass_reading_by_size,
    "Overall Passing Rate": pass_overall_by_size
})

# Insert Title for bins and index by them
scores_by_school_size.index.name = "Total Students"
scores_by_school_size = scores_by_school_size.reindex(group_names2)


# Format Data Frame
scores_by_school_size["Average Math Score"] = scores_by_school_size["Average Math Score"].map("{:.2f}".format)
scores_by_school_size["Percent Passing Math"] = scores_by_school_size["Percent Passing Math"].map("{:.2%}".format)
scores_by_school_size["Average Reading Score"] = scores_by_school_size["Average Reading Score"].map("{:.2f}".format)
scores_by_school_size["Percent Passing Reading"] = scores_by_school_size["Percent Passing Reading"].map("{:.2%}".format)
scores_by_school_size["Overall Passing Rate"] = scores_by_school_size["Overall Passing Rate"].map("{:.2%}".format)


scores_by_school_size
```

```python
# Scores by School Type
# Create new object with schools grouped by type 
by_type = education_df.groupby("type")

# Refigure averge scores within each school type
math_by_type = by_type["math_score"].mean()
reading_by_type = by_type["reading_score"].mean()
pass_math_by_type = education_df[students_df["math_score"] >= 70].groupby("type")["Student ID"].count()/by_type["Student ID"].count()
pass_reading_by_type = education_df[students_df["reading_score"] >= 70].groupby("type")["Student ID"].count()/by_type["Student ID"].count()
pass_overall_by_type = education_df[(students_df["math_score"] >= 70) & (education_df["reading_score"] >= 70)].groupby("type")["Student ID"].count()/by_type["Student ID"].count()

# Set-up Data Frame
scores_by_school_type = pd.DataFrame({
    "Average Math Score": math_by_type,
    "Percent Passing Math": pass_math_by_type,
    "Average Reading Score": reading_by_type,
    "Percent Passing Reading": pass_reading_by_type,
    "Overall Passing Rate": pass_overall_by_type
})

# Insert Title for index 
scores_by_school_type.index.name = "Type of School"



# Format Data Frame
scores_by_school_type["Average Math Score"] = scores_by_school_type["Average Math Score"].map("{:.2f}".format)
scores_by_school_type["Percent Passing Math"] = scores_by_school_type["Percent Passing Math"].map("{:.2%}".format)
scores_by_school_type["Average Reading Score"] = scores_by_school_type["Average Reading Score"].map("{:.2f}".format)
scores_by_school_type["Percent Passing Reading"] = scores_by_school_type["Percent Passing Reading"].map("{:.2%}".format)
scores_by_school_type["Overall Passing Rate"] = scores_by_school_type["Overall Passing Rate"].map("{:.2%}".format)


scores_by_school_type


```

Scores by School Type Analysis

The student performance data above, sorted by school type, tells an interesting story about who is "passing" and who is not. It misleads readers to believe that district schools are literally and figuratively failing their students.

Looking only at the Overall Passing Rate, it appears that district schools have a lot of catching up to do if only a little over half of their students are "passing" both core subjects whereas charter schools seem to be "passing" nearly all of their student body.  Make note of the 37-point difference in Overall Passing Rate.

The 70% cut-off for a passing score is a real detriment to the figures associated with district schools. They are underperforming, but the severity is exaggerated when reporting on percentage of students scoring in the passable range. In reality, district school students' math scores are averaging about 6.5 points below their charter school counterparts and their reading scores are averaging only 3 points behind.  

The difference in performance between students who attend charter schools and students who attend district schools is not a 37-point difference but rather, it's less than a 5-point difference.
